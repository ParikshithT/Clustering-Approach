# Task 1

Data sets,  “Data1.csv”, “Data2.csv”, “Data3.csv”, “Data4.csv”, “Data5.csv”, “Data6.csv”,
“Data7.csv”, “Data8.csv” contain the data points and their respective class information.
For each of the datasets follow the below steps
1. Use K-means and hierarchical clustering methods to generate clusters
2. Evaluate the performance of the clustering algorithm using external validation
metrics
3. Plot (2D or 3D) the data points for each dataset and color them according to the
original class
4. Plot (2D or 3D) the data points for each dataset and color them according to the
class allocated by the clustering algorithm

## Key Points about the approach

1. The number of Clusters chosen to run the algorithms for the datasets will depend on the number of Classes indicated in the actual data itself.
2. Since External validation is involved, which in itself a very rare case in clustering, there are no external validation metrics like f1 score or accuracy being used in this case because, the class labels will be chosen at random by the algorithm and so, these labels must be manually mapped to its corresponding label from the actual data, which is not an ideal situation because depending on the mapping, there will be different scores for the mertics.
3. The best possible way to evaluate the performance of the algorithms is to plot the data and visually make an assessment.
4. All the 3D plots are made interactive, anyone viewing the graph will be able to view the plots in different angles in three diemension.


## Clustering Algorithms

1. Both KMeans and Hierarchical Clustering Algorithms being used here are from the sklearn library.
2. The links to the libraries: [KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html/) and [Hierarchical](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html)
3. Reiterating that the method for calcuating distance for Agglomerative Clustering is Euclidean.

## Data 1

### 1.1 Reading the data

```python
# Import Data1.csv to a dataframe
import pandas as pd

data_1 = pd.read_csv('Data1.csv')
```

```python
data_1
```

```python
data_1.iloc[:,1:4].describe()
```

```python
print(data_1['X1'].isnull().value_counts())
print(data_1['X2'].isnull().value_counts())
print(data_1['X3'].isnull().value_counts())
```

There are no missing values in any of the variable columns

### 1.2 KMeans Clustering

```python
from sklearn.cluster import KMeans

data_1_mdl1 = KMeans(n_clusters=7, random_state = 4)
data_1_mdl1.fit(data_1.iloc[:,1:4])
```

```python
data_1['KMeans_Predicted_Class'] = list(data_1_mdl1.labels_)
```

### 1.3 Heirarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
data_1_mdl2 = AgglomerativeClustering(n_clusters=7)
data_1_mdl2.fit(data_1.iloc[:,1:4])
```

```python
data_1['Agglo_Predicted_Class'] = list(data_1_mdl2.labels_)
```

### 1.4 Results of Clustering

```python
data_1
```

```python
data_1['Class'] = data_1['Class'].astype('str')
data_1['KMeans_Predicted_Class'] = data_1['KMeans_Predicted_Class'].astype('str')
data_1['Agglo_Predicted_Class'] = data_1['Agglo_Predicted_Class'].astype('str')
```

#### 1.4.1 3D Plot of the actual data

```python
# Importing the library 
import plotly.express as px

# Plot
fig = px.scatter_3d(data_1,x='X1', y='X2',z='X3',
              color ='Class', title="Actual Data 1",
              width=750, height=500,
              template="plotly_dark")
fig.show()
```

#### 1.4.2 3D Plot of KMeans Clustering

```python
fig = px.scatter_3d(data_1,x='X1', y='X2',z='X3',
              color ='KMeans_Predicted_Class', title="KMeans Clustering",
              width=750, height=500,
              template="plotly_dark")
fig.show()
```

#### 1.4.3 3D Plot of Hierarchical Clustering

```python
fig = px.scatter_3d(data_1,x='X1', y='X2',z='X3',
              color ='Agglo_Predicted_Class', title="Hierarchical Clustering",
              width=750, height=500,
              template="plotly_dark")
fig.show()
```

### 1.5 Inferences from the plots

Visually Assessing both the plots we can conclude that:

Both the Algorithms have performed very well. All the points are classified correctly to their respective clusters without any misclassifications.

## Reading Data


```python
df = pd.read_csv('World Indicators.csv')
df
```

## Cleaning the data

Cleaning all the columns in the dataset, and making sure that all the values are represented in the correct format.


```python
df['GDP'] = df['GDP'].str.replace('$', "")
df['GDP'] = df['GDP'].str.replace(',', "")
df['Health Exp/Capita'] = df['Health Exp/Capita'].str.replace('$', "")
df['Health Exp/Capita'] = df['Health Exp/Capita'].str.replace(',', "")
df['Business Tax Rate'] = df['Business Tax Rate'].str.replace('%', "")

# Ensuring all the columns are in the right format.
df['GDP'] = df['GDP'].astype('float')
df['Health Exp/Capita'] = df['Health Exp/Capita'].astype('float')


df.head(20)
```

```python
type(df['GDP'][1])
```

## Approach 1

1. Remove all the columns with a large number of missing values. (Energy Usage and Lending Interest).
2. Rows with missing values are also removed.

### Removing Missing values

Check the number of missing values in each column

```python
df.isnull().sum()
```

We drop the columns 'Energy Usage'and 'Lending Interest', because it constains majority of null values or missing data.

```python
df1 = df.drop(['Energy Usage', 'Lending Interest'], axis = 1)
df1
```

Now, removing all the rows with missing values.

```python
df1.columns
```

```python
df1 = df1.dropna()
```

```python
df1.isnull().sum()
```

All the missing values are removed from the dataset. This is done because a missing rows form a minority part of about 10% - 20% of the data and we can still extract the right amount of intformation to proceed with the clustering algorithms. 

```python
df1.describe()
```

Checking for outliers in the dataset is essential to build an efficient model. 

```python
import seaborn as sns
```

As the above graph indicates, the maximum value of GDP is too large for the dataset and can lead to it having more 'weightage' or importance. In order to give equal importance to al the feature values, we scale or normalize the data. 

### Scaling the values

```python
from sklearn import preprocessing

Min_max = preprocessing.MinMaxScaler(feature_range = (0,1))

# cols = data.columns
df1['GDP'] = Min_max.fit_transform(df1[['GDP']])
df1['Days to Start Business'] = Min_max.fit_transform(df1[['Days to Start Business']])
df1['Health Exp/Capita'] = Min_max.fit_transform(df1[['Health Exp/Capita']])
df1['Hours to do Tax'] = Min_max.fit_transform(df1[['Hours to do Tax']])
df1['Life Expectancy Female'] = Min_max.fit_transform(df1[['Life Expectancy Female']])
df1['Life Expectancy Male'] = Min_max.fit_transform(df1[['Life Expectancy Male']])
df1['Mobile Phone Usage'] = Min_max.fit_transform(df1[['Mobile Phone Usage']])
df1['Business Tax Rate'] = Min_max.fit_transform(df1[['Business Tax Rate']])
```

```python
df1.head()
```

As the above output indicates, the values have been scaled to ensure consistency across tbe dataset.

### Checking for outliers 

```python
df1.hist(figsize=(15,10))
```

From the histograms above we can conclude that the dataset contains skewed values in quite a few features.

```python
import plotly.graph_objects as go

fig = go.Figure()


for column in df1.iloc[:,0:16]:
  fig.add_trace(go.Box(y=df1[column].values, name=df1[column].name))
  
fig.update_layout(template='plotly_dark')
fig.show()
```

<p>As the above boxplot indicates, there are many outliers in multiple columns.

```python
def impute_values(data):
  numeric_columns = list(df2.select_dtypes(include=[np.number]).columns.values)
  regions = list(data['Region'].unique())
  temp = pd.DataFrame()
  data3 = pd.DataFrame()
  for region in regions:
    temp = data[data['Region']== region]
    for col in numeric_columns:
      temp[col] = temp[col].fillna(temp[col].median())
    data3 = data3.append(temp, ignore_index=True)
  return data3
```
